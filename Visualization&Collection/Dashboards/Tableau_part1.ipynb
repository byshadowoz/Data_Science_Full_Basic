# Creating a Chart

1. Tables, here is where all the columns from the dataset are, here we can drag the columns to the canvas, Marks or Filters
2. Like Excel There are pages for the sheets and dashboards we can create new worksheets and new dashboards and also Stories
3. Here is where we can visualize our data in charts, there are a lots of types of plots, charts, graphs, we can also add a title (more details in the next module)
4. here we can visualize the columns and rows of the graph
5. With marks we can create labels, define the colors and size like plotly
6. Here we can se the legend of the graph
7. this will display a menu where we can select anothers graphs
