# Plots with Seaborn
Seaborn is a library based on matplotlib, it was designed for make more easy the graphs creation and make it more attractive and complex with less code 

```python
import seaborn as sns
```

## Datasets:

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
car_crashes = sns.load_dataset('car_crashes')
diamonds = sns.load_dataset('diamonds')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

# Plots:

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='time', size='size')
```

```python
sns.histplot(tips['tip'], bins=15)
```

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

```python
sns.barplot(x='cut', y='price', data=diamonds)
```

```python
sns.boxplot(x='sex', y='age', data=titanic, hue='class')
```

```python
sns.stripplot(x='sex', y='age', data=titanic, hue='class', dodge=True)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex')
```

```python
sns.pairplot(iris, hue='species')
```

## Heat Map
I separated this plot from the others as this plot is a much more important one when it comes to finding the best relationships between data being useful both in data visualization and machine learning

```python
titanic.select_dtypes(['number']).corr()
```

```python
diamonds.select_dtypes(['number']).corr()
```

```python
sns.heatmap(titanic.select_dtypes(['number']).corr(), annot=True)
```

```python
sns.heatmap(diamonds.select_dtypes(['number']).corr(), annot=True)
```

# Customizing plots
let's add some colors
more colors: https://seaborn.pydata.org/tutorial/color_palettes.html

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='time', size='size', palette='Spectral')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex', cmap='coolwarm')
```

```python
sns.stripplot(x='sex', y='age', data=titanic, hue='class', dodge=True, palette='YlGnBu')
```

```python
sns.heatmap(planets.select_dtypes(['number']).corr(), annot=True, cmap='coolwarm')
```
