# Plots in Python with Matplotlib

Before we create Plots to find insights and patters, Like the machine learning models we first need to see the Data for analise, clean it and explore the columns, what the columns means and what is the important or importants Labels and/or scale the data. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly as pt
```

```python
df = pd.read_csv('user_behavior_dataset.csv', index_col=0)#data from https://www.kaggle.com/datasets/valakhorasani/mobile-device-usage-and-user-behavior-dataset
```

```python
df
```

Let's check what the labels means:

- User ID: Unique identifier for each user.
- Device Model: Model of the user's smartphone.
- Operating System: The OS of the device (iOS or Android).
- App Usage Time: Daily time spent on mobile applications, measured in minutes.
- Screen On Time: Average hours per day the screen is active.
- Battery Drain: Daily battery consumption in mAh.
- Number of Apps Installed: Total apps available on the device.
- Data Usage: Daily mobile data consumption in megabytes.
- Age: Age of the user.
- Gender: Gender of the user (Male or Female).
- User Behavior Class: Classification of user behavior based on usage patterns (1 to 5).

# Scatter Plot:


### we define X and Y just like machine learning but in this case X would be just one column
## Let's make an example using screen time and battery drain columns

```python
x = df['Battery Drain (mAh/day)']
y = df['Screen On Time (hours/day)']

plt.scatter(x, y)
plt.show() # this method is to see the graph
```

Now we add some methos to improve our graph visualize

```python
x = df['Battery Drain (mAh/day)']
y = df['Screen On Time (hours/day)']

plt.title('Screen Time & batery Drain Graph')
plt.xlabel('Battery Drain(mAh/day')
plt.ylabel('Screen On Time(hours/day)')
plt.scatter(x, y, color='red')
plt.show()
```

Let's see with others values

```python
x = df['Number of Apps Installed']
y = df['App Usage Time (min/day)']

plt.title('Relation between App usage and Number of apps installed')
plt.xlabel('Number of Apps installed')
plt.ylabel('Total Apps Usage Time (min/day)')
plt.scatter(x, y, color='blue')
plt.show()
```

```python
x = df['Number of Apps Installed']
y = df['Data Usage (MB/day)']

plt.title('Number of Apss Installed and Data usage')
plt.xlabel('Number of Apps installed')
plt.ylabel('Data Usage (MB/day)')
plt.scatter(x, y, color='purple')
plt.show()
```

## Also we can make a way to do this more compact and organized using subplots():

```python
fig, ax = plt.subplots(
    figsize=(7,7), nrows=2, ncols=2
)
fig.suptitle('Users behavior')

total_ax = [0,1,2,3]
colors=['red','green','blue','purple']

ax = fig.axes[total_ax[0]]
ax.set_xlabel('Battery Drain (mAh/day)')
ax.set_ylabel('Screen On Time (Hours/day)')
ax.scatter(df['Battery Drain (mAh/day)'], df['Screen On Time (hours/day)'], color=colors[0], marker='+', linewidths=1)


ax = fig.axes[total_ax[1]]
ax.set_xlabel('Number of Apps Installed')
ax.set_ylabel('Screen On Time (Hours/day)')
ax.scatter(df['Number of Apps Installed'], df['Screen On Time (hours/day)'], color=colors[1], marker='.')


ax = fig.axes[total_ax[2]]
ax.set_xlabel('Number of Apps Installed')
ax.set_ylabel('App Usage Time (min/day)')
ax.scatter(df['Number of Apps Installed'], df['App Usage Time (min/day)'], color=colors[2], marker='o')


ax = fig.axes[total_ax[3]]
ax.set_xlabel('Number of Apps Installed')
ax.set_ylabel('Data Usage (MB/day)')
ax.scatter(df['Number of Apps Installed'], df['Data Usage (MB/day)'], color=colors[3], marker='x')
```

# Pie Charts:

```python
df = pd.read_csv('user_behavior_dataset.csv', index_col=0)
df_filtred = df[df['User Behavior Class'] > 3]
filter=df_filtred['Device Model'].unique()

x = []
for v in filter:
    df_for = df_filtred[df_filtred['Device Model'] == v]
    x.append(df_for['Device Model'].count())

x
```

```python
plt.title('percentage of users')
plt.pie(x, labels=filter, autopct='%.2f%%', explode=[0.05,0.01,0.1,0.2,0])
```

# Line plots:

```python
df = pd.read_csv('netflix_titles.csv') # data from: https://www.kaggle.com/datasets/shivamb/netflix-shows
df
```

```python
years = df['release_year'].unique()
counts = []
for y in range(0, len(years)):
    count_year = df[df['release_year'] == years[y]]
    counts.append(count_year.count().max())
df_count_years = pd.DataFrame({'Year': years,
                              'Number_of_movies': counts})
df_count_years.sort_values(ascending=True, by='Year', inplace=True)
```

```python
plt.title('Number of movies throught the years')
plt.plot(df_count_years['Year'], df_count_years['Number_of_movies'], linestyle='--', color='black')
```

# 3D Plots:

```python
ax = plt.axes(projection='3d')
x = df['BMI']
y = df['Age']
z = df['Fat_Percentage']
ax.scatter(x,y,z, color='cyan', marker='+')
ax.set_xlabel('BMI')
ax.set_ylabel('Age')
ax.set_title('3d_plot | BMI, Age, Fat% relation')
```

```python
ax = plt.axes(projection='3d')

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap='Spectral')
```

```python
df = pd.read_csv('disney_movies_total_gross.csv') # data from: https://www.kaggle.com/datasets/rashikrahmanpritom/disney-movies-19372016-total-gross
df['release_date'] = pd.to_datetime(df['release_date'])
df.dropna(inplace=True)

df['inflation_adjusted_gross'] = df['inflation_adjusted_gross'] // 1000000
df['inflation_adjusted_gross']

income_labels = list(range(0, 6000, 400))
plt.yticks(income_labels, [f'{x}m $' for x in income_labels])
plt.title('Disney gross movies (inflation adjusted) in USD', fontsize=20)
plt.xlabel('years')
plt.plot(df['release_date'].head(10), df['inflation_adjusted_gross'].head(10), c='green')
```

```python
income_labels = list(range(0, 1000, 50))
plt.yticks(income_labels, [f'{x}m $' for x in income_labels])
plt.title('Disney gross movies (inflation adjusted) in USD', fontsize=20)
plt.xlabel('years')
plt.plot(df['release_date'].tail(100), df['inflation_adjusted_gross'].tail(100), c='red')
```
