# Association Rules
***

Association rules is a machine learning algorithm wich is widely used in Markets, retailers, recomendation systems etc.
To uncover assosiation between items.
This algorithm is based on a Rule which is (IF A THEN B) that means if the customer bought the product A then wich is the chances of product B being bought. and find the chance of other customer repeat the same pattern

***
- ## **Apriori Algorithm:**
It proceeds by identifying the frequent individual items in the database and extending them to larger and larger item sets as long as those item sets appear sufficiently often in the database. The frequent item sets determined by Apriori can be used to determine association rules which highlight general trends in the database: this has applications in domains such as market basket analysis.
### Key Concepts
- Itemsets: Sets of items (e.g., products, attributes) that co-occur in a transaction or record.
- Support: The percentage of transactions or records in which an itemset appears.
- Confidence: The percentage of transactions or records in which an itemset appears, given the presence of a subset itemset.

## Let's make an example:

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel('Online_retail.xlsx')
df.head()
```

### Clean data:

```python
df.info()
```

```python
df.isna().any()
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
df.info()
```

### Let's use only the sales in France to keep the dataset small

```python
basket = (df[df['Country'] == 'France'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0)
         .set_index('InvoiceNo'))
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.map(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules.head()
```

```python
rules[(rules['lift'] >= 6) & (rules['confidence'] >= 0.8)]
```
